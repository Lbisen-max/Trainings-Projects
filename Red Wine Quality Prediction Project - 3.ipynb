##### Checking corelation between depndenet and indepndenet features and among the indepndenet features.

```python
corrrelation = wine.corr()
```

```python
plt.figure(figsize=(16,16))
sns.heatmap(corrrelation,cbar=True, square=True,fmt='.2g',annot=True,annot_kws={'size':8},cmap='Blues')
```

We can see there are feature like chlorides ,free sulfur dioxide ,total sulfur dioxide and pH showing negatively corelated.
fixed acidity and chlorides showing highly corelated.we will check with other method alos to make sure if they are relly corelated.

Checking if there is any "0"value in our features and we we replacing by using mean.

```python
wine['volatile acidity'] = wine['fixed acidity'].replace(0,wine['fixed acidity'].mean())
wine['citric acid'] = wine['citric acid'].replace(0,wine['citric acid'].mean())
wine['chlorides'] = wine['chlorides'].replace(0,wine['chlorides'].mean())
wine['free sulfur dioxide'] = wine['free sulfur dioxide'].replace(0,wine['free sulfur dioxide'].mean())
wine['total sulfur dioxide'] = wine['total sulfur dioxide'].replace(0,wine['total sulfur dioxide'].mean())
```

### Outlianrs Detection And Removel

```python
plt.figure(figsize=(16,16))
ploatnumber = 1
for i in wine:
    if ploatnumber <=12:
        plt.subplot(4,3,ploatnumber)
        sns.boxplot(wine[i],orient='v')
        plt.xlabel(i,fontsize=10)
        ploatnumber+=1
plt.show()
```

Importing "zscore" for outliar removel

```python
from scipy.stats import zscore
```

```python
z_score = zscore(wine[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','sulphates','alcohol']])
abs_z_score = np.abs(z_score)
filtering_entry = (abs_z_score <3).all(axis=1)
wine = wine[filtering_entry]
wine.describe()
wine.head()
```

```python
wine.shape
```

#### Feature Engeneering !

* Importing standard scaler and VIF for multicollinearity

```python
from sklearn.preprocessing import StandardScaler #### For normalize the data
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

### label binarizations

* Divde the labels and features and converting the label data into binory as we have to predit the quility of wine
  hence we can assume that if value is 7 or 8 then it indicates a good quality of wine if less then 7 then wine quality is not good.

```python
X = wine.drop(columns=['quality'],axis=1)
y = wine['quality'].apply(lambda y_values: 1 if y_values>=7 else 0) #label binarizations

print("Shape of X =>",X.shape)
print("Shape of y =>",y.shape)
```

```python
y.value_counts()
```

Scaleing the data

```python
scaler = StandardScaler()
X_scaler=scaler.fit_transform(X)
```

```python
we can see VIP showing no multicollinearity how ever 
```

```python
vif = pd.DataFrame()
vif['score'] = [variance_inflation_factor(X_scaler,i) for i in range (X_scaler.shape[1])]
vif['feature'] = X.columns
vif
```

As check VIF scores we can see there not much effect of multicollinearity bwteenc columns.but in fixed acidity and volatile acidity
scores showinh inf hence removed both and rechecked again VIF scores not much different found hence.so we can keep as it is thease columns.


```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scaler,y,test_size=0.25,random_state=111)
```

```python
print("X_train size of data",X_train.shape)
print("X_test size of data",X_test.shape)
print("y_train size of data",y_train.shape)
print("y_train size of data",y_test.shape)
```

## Model Building

Importing libraries for model building......!

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
lr=LogisticRegression()
knn=KNeighborsClassifier()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
adb=AdaBoostClassifier()

print("Model is created")
```

```python
lr.fit(X_train, y_train)
knn.fit(X_train,y_train)
dt.fit(X_train,y_train)
rf.fit(X_train,y_train)
adb.fit(X_train,y_train)

print("Model is trained")
```

```python
print("Lr classification score",lr.score(X_train,y_train))
print("knn classification score",knn.score(X_train,y_train))
print("dt classification score",dt.score(X_train,y_train))
print("rf classification score",rf.score(X_train,y_train))
print("adb classification score",adb.score(X_train,y_train))
```

```python
lr_yprad = lr.predict(X_test)
knn_yprad = knn.predict(X_test)
dt_yprad = dt.predict(X_test)
rf_yprad = rf.predict(X_test)
adb_yprad = adb.predict(X_test)
```

### HyperParameter Tuning !

```python
from sklearn.model_selection import GridSearchCV
```

```python
lr.get_params().keys()  # to check the parameters
```

```python
parm_grid = {'penalty':['l1', 'l2'],
            'dual':[True,False],'tol':[True,False],
            'fit_intercept':[True,False],'intercept_scaling':[True,False],
            'solver':['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga']}
print('parm_grid',parm_grid)
```

```python
gridsearch = GridSearchCV(lr, param_grid = parm_grid , cv=5)
```

```python
gridsearch.fit(X_train, y_train)
```

```python
gridsearch.best_params_
```

```python
print(f'train accuracy','=>',gridsearch.score(X_train,y_train))
print(f'test accuracy','=>',gridsearch.score(X_test,y_test))
```
